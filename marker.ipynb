```
# Imports
import folium
import random
import pandas as pd
from scipy.spatial import ConvexHull
from h3 import h3

df = pd.read_csv("marugame_poi.csv",encoding="shift-jis")
print(df)
```

```
m = folium.Map(location=[35.650273, 139.708349],zoom_start=10)
for lat, lon ,name,area,status,parking in zip(df.latitude, df.longitude, df.name,df.area,df.status,df.parking):
  if status == "単独":
    folium.Marker(location = [lat, lon],popup =area,tooltip = name,icon=folium.Icon(color="blue")).add_to(m)
  else:
    folium.Marker(location = [lat, lon],popup =area,tooltip = name,icon=folium.Icon(color="black")).add_to(m)

folium.Marker(location = [35.72630738, 139.5991181],popup ="丸亀製麺 上石神井店",tooltip = "丸亀製麺 上石神井店",icon=folium.Icon(color="red")).add_to(m)
folium.Marker(location = [35.71826597, 139.4373937],popup ="丸亀製麺 立川若葉店",tooltip = "丸亀製麺 立川若葉店",icon=folium.Icon(color="red")).add_to(m)
folium.Marker(location = [35.74546194, 139.4099606],popup ="丸亀製麺 武蔵村山店",tooltip = "丸亀製麺 武蔵村山店",icon=folium.Icon(color="red")).add_to(m)
display(m)
```
